```python
import pickle
import pandas as pd
```

```python
lr = pickle.load(open('/scratch/res2/logreg_model_v1', 'r'))
fvs = pd.read_csv('/scratch/res2/candset_fvs_v2.csv')
```

```python
coeff = list(lr.coef_[0])
intercept = lr.intercept_[0] 
```

```python
feature_attrs = list(fvs.columns)
feature_attrs.remove('ltable.ID')
feature_attrs.remove('rtable.ID')
proj_fvs = fvs[feature_attrs]
```

```python
matches = 0
max_sc = 0
max_f=-100000
for row in proj_fvs.itertuples(index=False):
    score = 0
    for i in xrange(170):
        score += coeff[i]*row[i]
        if row[i] > max_f:
            max_f = row[i]
        #print coeff[i]*row[i]
    score += intercept
    if score > max_sc:
        max_sc = score
    if score >= 0:
        matches += 1
matches, max_sc, max_f
```
